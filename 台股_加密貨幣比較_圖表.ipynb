# 以太幣

```python
ETH = yf.Ticker('ETH-USD').history(period='max')          #以太幣 (已經轉換成台幣了)
ETH = ETH[['Close']] * 30.95    #先換成台幣的價格(11/28的價格)
ETH.columns = ['price']
ETH['ROI'] = ETH.price.pct_change()
ETH['price'] = round(ETH['price'],2)
ETH
```

# 比特幣

```python
BTC = yf.Ticker('BTC-USD').history(period='max')     #比特幣 (台幣)
BTC = BTC[['Close']] * 30.95
BTC.columns = ['price']
BTC['ROI'] = BTC.price.pct_change()     
BTC['price'] = round(BTC['price'],2)
BTC = BTC['2017-11-09':]                     #改從2017-11-09開始算，配合ETH數據筆數
BTC
```

```python
BTC.ROI.describe()
```

```python
BTC.ROI.plot()
```

# 比特幣與以太幣 起伏比較

```python
plt.plot(BTC.index,BTC['ROI'],label = 'BTC')
plt.plot(ETH.index,ETH['ROI'], color='red', label="ETH")
plt.xlabel('Date', fontsize="10") # 設定 x 軸標題內容及大小
plt.ylabel("ROI", fontsize="10",rotation=True) # 設定 y 軸標題內容及大小

plt.ylim(-0.5,0.3)       #y-axis範圍設定一萬~一百萬，設太大ETH起伏會更不明顯

plt.title('BTC and ETH ROI comparison', fontsize="18") # 設定圖表標題內容及大小

plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
y = BTC['ROI']
x = sm.add_constant(ETH['ROI'])    #用以太幣去預測比特幣
result = sm.OLS(y,x,missing='drop').fit(cov_type='HAC',cov_kwds={'maxlags':5})     #做優化
print(result.summary())
```
