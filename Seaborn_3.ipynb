```python
flights = sns.load_dataset('flights')
flights
```

```python
flights.info()
```

```python
pflights = flights.pivot_table(values='passengers',
 index='year',columns='month')
```

```python
sns.heatmap(pflights)
```

```python
sns.heatmap(pflights,annot=True)
```

```python
sns.heatmap(pflights,annot=True,fmt='d')
```

```python
sns.heatmap(pflights,annot=True,fmt='d',
 linecolor='white',linewidths='0.1')
```

```python
sns.clustermap(pflights,annot=True,fmt='d')
```

```python
sns.clustermap(pflights,standard_scale=1,annot=True)
# standard_scale針對column(1)或row(0)標準化
```
