1. displot()集合了matplotlib的hist()与核函数估计kdeplot的功能，
   增加了rugplot分布观测条显示与利用scipy库fit拟合参数分布的新颖用途

```python
sns.distplot(tips['total_bill'])
```

1. 針對單變數觀察值
2. 常用於快速查看單變數分佈
3. 劃出Histogram
4. 適配核密度估計值(Kernel Density Estimate, KDE)

```python
sns.distplot(tips['total_bill'],kde=False,rug=True)
```

1. 去除密度曲線、添加rugplot (在每個觀察值處畫一個垂直刻度)
2. kdeplot(核密度估计图, kernel density estimation): 
   是在概率论中用来估计未知的密度函数，属于非参数检验方法之一。
   通过核密度估计图可以比较直观的看出数据样本本身的分布特征
3. rug：控制是否生成观测数值的小细条

```python
sns.distplot(tips['total_bill'],kde=False,rug=True,bins=20)
```

bins()：int或list，將直方圖做一個劃分

```python
sns.distplot(tips['total_bill'],hist=False)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip',data=tips,kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip' ,data = tips,kind='reg')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip' ,data = tips,kind='resid')   # residuals殘差
```

```python
# KDE，等高線圖(contour plot)
sns.jointplot(x = 'total_bill',y = 'tip' ,data = tips,height=10,kind = 'kde')
```

```python
# Hexbin plot
# 二維六邊形圖(Hexagonal Bin Plot)
sns.jointplot(x = 'total_bill',y = 'tip' ,data = tips,height=10,kind = 'hex')
```
